# Exercise 1.A NumPy: Manipulating Arrays

```python
import numpy as np
```

## Looping Through Arrays

```python

arrA = np.array([[11, 12, 13],
                 [14, 15, 16]])
print(arrA)
```

```python

for x in arrA:
    print(x * 2)
```

```python

for x in arrA:
    for y in x:
        print(y * 2)
```


The first loop goes through rows (arrays) and multiplies entire rows by 2.
The second loop goes through each individual value and multiplies each element by 2 one at a time.


## Joining and Stacking Arrays

```python

joined_array_1 = np.concatenate((arrA, arrA))
print(joined_array_1)
```

```python

joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)
```

```python

stacked_array_1 = np.stack((joined_array_1, joined_array_1))
print(stacked_array_1)
```

```python

try:
    stacked_array_2 = np.stack((joined_array_1, joined_array_2))
    print(stacked_array_2)
except Exception as e:
    print(e)

# Error happens because arrays have different shapes and cannot be stacked.
```

## Splitting Arrays

```python

splitsky = np.array_split(joined_array_1, 2)
print(splitsky)

# This creates 2 arrays, each with the same number of columns.
```

```python

pliterino = np.array_split(joined_array_1, 3)
print(spliterino)

# This creates 3 arrays; sizes may not be equal.
```
