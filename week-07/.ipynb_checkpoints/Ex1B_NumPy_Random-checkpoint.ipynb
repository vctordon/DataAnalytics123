
## Generating Random Numbers


```python
r_1 = np.random.randint(1, 21, size =2)
print(r_1)
```

```python
r_2= np.random.randint(1,21, size=(1, 2))
print(r_2)
```

```python
r_3 = np.random.rand(2, 1)
print(r_3)
```

```python
r_4 = np.random.randint(20, 101, size=(4, 5))
print(r_4)
```

```python
r_5 = np.random.rand(4, 5)
print(r_5)
```

## Generating Random Choices from an Array

```python

students = np.array([
    "Alexus","Naomi","Alondra","Daniel","Corey","Lorah",
    "Blake","Luis","Charles","Sonia","Hana","Smaya"
])
print(students)
```

```python
call_on = np.random.choice(students)
print(call_on)
```

```python

call_on = np.random.choice(students)
print(call_on)
```

```python

call_on = np.random.choice(students)
print(call_on)
```

```python
breakout_groups = np.random.choice(students, size=(3, 4))
print(breakout_groups)

# This may not be effective because the same student can appear multiple times
# and some students might not be select
```

```python

breakout_groups = np.random.choice(students, size=(3, 4))
print(breakout_groups)
```

```python

probabilities = [0.1, 0.1, 0.05, 0.05, 0.1, 0.1, 0.3, 0.0, 0.05, 0.05, 0.05, 0.05]

call_on_weighted = np.random.choice(students, size=5, p=probabilities)
print(call_on_weighted)
```


## Generating a Normal Distribution


```python

bell_curve = np.random.normal(loc=10, scale=5, size=(4, 5))
print(bell_curve)
```
