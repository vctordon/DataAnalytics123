# Exercise 1.A NumPy: Manipulating Arrays


```python
import numpy as np
```

# Looping Through Arrays

```python
# Create arrA 
arrA = np.array([[11,12,13], [14,15,16]])
print(arrA)
```

```python
# Loop through the first dimension
for x in arrA:
    print(x * 2)
```

```python
# Loop through the second dimension 
for x in arrA:
    for y in x:
        print(y * 2)
```

# The first loop does the function from left to right vs the second from top to bottom

```python
# Loop through second dimension with loop_result
for x in arrA:
    for y in x: 
        loop_result = x * 2
        print(loop_result)

# Prints tow multiplied by 2 repeated because X (row) is used instead of y (the individual element)
```

## Joining and Stacking Arrays

```python
# Concatenate on axis 0
joined_array_1 = np.concatenate((arrA, arrA))
print(joined_array_1)
```

```python
# Concatenate on axis 1
joined_array_2 = np.concatenate((arrA, arrA),axis =1)
print(joined_array_2)
```

```python
# Stack arrays (default axis=0)
stacked_array_1 = np.stack((joined_array_1, joined_array_1))
print(stacked_array_1)
```

```python
# Attempt stacking mismatched shapes
stacked_array_2 = np.stack((joined_array_1, joined_array_2))
print(stacked_array_2)

# Error occurs because the arrays have different shapes and cannot be stacked together
```

```python

# Concatenate joined_array_1 with arrA
another_joined_array = np.concatenate((joined_array_1, arrA))
print(another_joined_array)
```

```python

# Concatenate on axis 1 (will error)
another_2 = np.concatenate((joined_array_1, arrA), axis=1)
print(another_2)

# Error occurs because the number of rows does not match for axis=1 concatenation.
```


# Using stacking helper functions
h_stack = np.hstack((arrA, arrA))  # stacks horizontally (side by side)
v_stack = np.vstack((arrA, arrA))  # stacks vertically (on top)
d_stack = np.dstack((arrA, arrA))  # stacks depth-wise (adds a third dimension)

print(h_stack)
print(v_stack)
print(d_stack)



## Splitting Arrays


```python

# Split into 2
splitsky = np.array_split(joined_array_1, 2)
print(splitsky)

# This creates 2 arrays, each still 2D.
```

```python

# Split into 3
spliterino = np.array_split(joined_array_1, 3)
print(spliterino)

# This creates 3 arrays; they are still 2D but may not be evenly sized.
```


## Searching and Replacing Array Values


```python

# Use where() for even numbers
result_1 = np.where(arrA % 2 == 0, arrA, 'odd')
print(result_1)
```

## Sorting and Filtering Arrays


```python

# Create arrB and sort
arrB = np.array([[7, 6], [13, 5], [7, 3]])
arrB_sorted = np.sort(arrB)
print(arrB_sorted)
```

```python

# Filter values less than 7
print(arrB)

z = arrB < 7
arrC = arrB[z]

print(arrC)
```

```python

# Filter values greater than or equal to 7
y = arrB >= 7
arrD = arrB[y]

print(arrD)
```
