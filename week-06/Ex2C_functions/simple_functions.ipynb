```python
def favorite_things(name,movie,music):
    return f"My name is {name}.\nMy favorite movie is {movie}.\nMy favorite artist is {music}"
```

```python
print(favorite_things("Victoria","Shrek","Ariana Grande"))
```

```python
def why_im_here(purpose):
    return purpose
```

```python
print(why_im_here("I joined YUU to become a Data Analyst."))
```

```python
def favorite_place(place,why):
    return f"One of my favorite places is {place} because {why}."
```

```python
print(favorite_place("Tenessee", "I love the mountains."))
```

```python
print(favorite_place("New York", "I love bagels and the big apple"))
```
