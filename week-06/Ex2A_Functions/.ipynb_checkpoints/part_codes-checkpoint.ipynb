# Exercise 2.A Using built-in

```python
def get_supplier_code(code):
    for i in range(len(code)):
        if code[i] == ':':
            return code[:i]

#get the supplier
```

```python
def get_product_num(code):
    start = code.find(':') + 1
    end = code.find('-')
    return code[start:end]

#get the product number
```

```python
def get_size(code):
    return code.split('-')[1]
#get the size
```

```python
code1 = "ACME:123-L"
code2 = "DI:12345-M"
code3 = "ACE:1-12"
#storing the example codes
```

```python
supplier = get_supplier_code(code1),
product = get_product_num(code1),
size = get_size(code1)

print(f"For {code1}, the supplier is {supplier}. Part number {product} is available in size {size}.")
```
