In this lab, we are working with Python datetime functions.

```python
import datetime
```

# Today 

```python
today = datetime.datetime.now()
today
```

```python
formatted_today = today.strftime("Today is %A, %B %d, %Y")
print(formatted_today)
```

# My birthday

```python
my_birthday = datetime.date(1997,5,21)
```

```python
print("My birthday is", my_birthday.strftime("%m/%d/%y"))
```

# next 90 days  

```python
ninety_d = today + datetime.timedelta(days=90)
```

```python
print("90 days from today is", ninety_d.strftime("%B %d, %Y"))
```

# time for dinner

```python
dinner_time = datetime.time(19,30)
```

```python
print("Let’s meet for dinner at", dinner_time.strftime("%I:%M %p"))
```
